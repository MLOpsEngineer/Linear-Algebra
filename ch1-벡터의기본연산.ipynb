```python
import numpy as np

#내적 분배 법칙
a = np.array([1,2,3])
b = np.array([3,5,9])
c = np.array([12,0,1])

res1 = np.dot(a, b+c)
res2 = np.dot(a,b) + np.dot(a,c)

print(res1, res2)
```
